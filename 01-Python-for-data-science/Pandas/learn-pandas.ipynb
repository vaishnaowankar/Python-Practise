# Creating dataframes from python dictionaries



```python
people = {
    "first" : ['Vaishnao','Avyukt','Reyansh'],
    "last" : ['Wankar','Harjai','Lamba'],
    "email" : ['vaishnaowankar@gmail.com','avyuktharjai@gmail.com','reyanshlamba@gmail.com']
}
```

```python
# we know we can access any key in dict
people['email']
```

```python
# Creating a DataFrame
python_df = pd.DataFrame(people)
```

```python
python_df
```

```python
# we can similarly access any column like python dict from dataframe
python_df['email']
```

```python
# we can also use dot method to do the same but is not prefereed because because the attributes may be same as pandas methods , hence would create a problem.

python_df.email

# PRoblem : suppose if DataFrame contains a column names count then accessing it using dot method will call the pandas count method .
```

# Some more dataframe access methods

```python
python_df[['first','email']]
```

```python
# access the columns list
python_df.columns
```

```python
# to access rows we have two methods
# iloc - access the location by index
# loc - accessing the location by labels

python_df.iloc[[0,1],2]
# the sytax is first index has rows index , the second index is the column index for specified rows
```

```python
# for only rows
python_df.iloc[[0,1]]
# returns the rows at index 0 and 1
```

```python
# using loc

python_df.loc[[0,1],['email','first']]
# we can also specify multiple index for columns in iloc
# loc looks similar to iloc but we will see the difference while dealing with the above stackoverflow data
```

# Filtering the dataset


### Working with our demo data set (dict) first


```python
python_df['last'] == 'Harjai'
```

```python
# better method to do it
filt = (python_df['last'] == 'Harjai')
```

```python
# returns the DataFrame 
python_df[filt]
```

```python
# Even better method is to use loc
python_df.loc[filt,'email']
```

```python
filt2 = (python_df['last'] == 'Harjai') & (python_df['first'] == 'Avyukt')
```

```python
python_df[filt2]
```

```python
filt3 = (python_df['first'] == 'Vaishnao') | (python_df['last'] == 'Lamba')
```

```python
python_df[filt3]
```

```python
# Getting output opposite to that of the filter
python_df.loc[~filt3 , 'email']

# using the ~ sign
```
